```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
#$ tshark -r sample-cap.pcap -Tfields -E header=y -E separator=';' -e frame.number -e frame.time -e frame.len -e frame.protocols -e ip.proto -e ip.src -e tcp.srcport -e udp.srcport -e ip.dst -e tcp.dstport -e udp.dstport > sample-cap.csv 

# CSVファイルをdataframeにできます。
dframe = pd.read_csv('sample-cap.csv', sep=";")
```

```python
#dframe.head()
dframe[dframe['frame.protocols'].str.contains ('udp')]
```

```python

# 必要なモジュールをインポートします。
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Protocol
sns.countplot('frame.protocols',data=dframe)
```

```python
def proto(df):
    # データを取得します。
    prt = df
    if prt.count('udp'):
        return 'udp'
    elif prt.count('tcp'):
        return 'tcp'
    else:
        return prt

#protocolという新しい列を追加します。
dframe['protocol'] = dframe['frame.protocols'].apply(proto)
#dframe['frame.protocols'].apply(proto)
```

```python
dframe.head()
```

```python
sns.countplot('protocol',data=dframe)
```

```python
# クロス集計表が作れます。
dframe2 = pd.crosstab(dframe.protocol,dframe['frame.len'],margins=True)
```

```python

# pandasで簡単にヒストグラムを描く方法です。
dframe['frame.len'].hist(bins=70)

#sns.barplot(x='sex', y='survived', data=titanic, hue='class')
#sns.barplot(x='frame.len', y='protocol', data=dframe)
#dframe2['frame.len']
```

```python

# FacetGridを使うと、複数のカーネル密度推定のグラフを1つのプロットに描くことができます。

# protocolで層別化して、グラフを少し横長に設定します。
fig = sns.FacetGrid(dframe, hue="protocol",aspect=4)
# mapを使って、性別ごとにkdeplotを描くようにします。
fig.map(sns.kdeplot,'frame.len',shade= True)

# x軸の範囲を設定します。
fig.set(xlim=(0,1500))

# 凡例を付け加えておきましょう。
fig.add_legend()
```
